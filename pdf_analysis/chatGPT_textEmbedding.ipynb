```python
def get_gutenberg(url):
    loader = GutenbergLoader(url)
    data = loader.load()
    return data
```

```python
wellness_data = get_gutenberg('https://www.gutenberg.org/cache/epub/33917/pg33917.txt')

text_splitter = TokenTextSplitter(chunk_size=1000, chunk_overlap=0)
wellness_doc = text_splitter.split_documents(wellness_data)

embeddings = OpenAIEmbeddings()
vectordb = Chroma.from_documents(wellness_doc, embeddings)
#                                  , persist_directory=persist_directory)
# vectordb.persist()
```
